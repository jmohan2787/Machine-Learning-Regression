```python
import pandas as pd
```

```python
data=pd.read_csv('Social_Network_Ads.csv')
```

```python
data
```

```python
data=data.drop('User ID',axis=1)
```

```python
data
```

```python
datanew=pd.get_dummies(data,drop_first=True)
```

```python
datanew.shape
```

```python
datanew.columns
```

```python
Independant=datanew[['Age', 'EstimatedSalary','Gender_Male']]
```

```python
dependant=datanew[['Purchased']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(Independant,dependant,test_size=1/3,random_state=0)
```

```python
from sklearn.svm import SVC
```

```python
classifier=SVC(kernel='rbf',random_state=0)
```

```python
classifier.fit(X_train,Y_train)
```

```python
y_predict=classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm=confusion_matrix(Y_test,y_predict)
```

```python
cm
```

```python
from sklearn.metrics import classification_report
```

```python
classify_report=classification_report(Y_test,y_predict)
```

```python
print(classify_report)
```
