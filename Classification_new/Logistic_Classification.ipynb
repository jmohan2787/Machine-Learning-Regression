```python
import pandas as pd
```

```python
data=pd.read_csv("Social_Network_Ads.csv")
```

```python
data
```

```python
datanew=pd.get_dummies(data,drop_first=True)
```

```python
datanew['Purchased'].value_counts()
```

```python
datanew
```

```python
datanew=datanew.drop("User ID",axis=1)
```

```python
datanew.columns
```

```python
Independand=datanew[['Age', 'EstimatedSalary','Gender_Male']]
```

```python
depenedand=datanew[['Purchased']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(Independand,depenedand,test_size=1/3,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
classifier=LogisticRegression(random_state=0)
```

```python
classifier.fit(X_train,Y_train)
```

```python
y_predict=classifier.predict(X_test)
```

```python
y_predict
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm=confusion_matrix(Y_test,y_predict)
```

```python
cm
```

```python
from sklearn.metrics import classification_report
```

```python
report=classification_report(Y_test,y_predict)
```

```python
report
```
