```python
import pandas as pd
data=pd.read_csv('CKD.csv')
```

```python
datanew=pd.get_dummies(data,drop_first=True)
```

```python
datanew
```

```python
datanew.columns
```

```python
independand=datanew[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
       'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
       'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
       'appet_yes', 'pe_yes', 'ane_yes']]
```

```python
dependand=datanew[['classification_yes']]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(independand,dependand,test_size=1/3,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
scTrain=sc.fit_transform(X_train)
scTrest=sc.fit(X_test)
```

```python
#criterion{“gini”, “entropy”, “log_loss”},splitter{“best”, “random”},max_featuresint, float or {“sqrt”, “log2”}, default=None
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
params_grid=param_grid = {'kernel':['linear','rbf','poly','sigmoid'],
             'gamma':['auto','scale'],
             'C':[10,100,1000,2000,3000]}
grid=GridSearchCV(SVC(probability=True), params_grid, refit=True, verbose=3, n_jobs=-1,scoring='f1_weighted')
grid.fit(X_train,Y_train)
```

```python
cv=grid.cv_results_
table=pd.DataFrame.from_dict(cv)
```

```python
table
```

```python
y_predict=grid.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(Y_test,y_predict)
cm
```

```python
from sklearn.metrics import classification_report
report=classification_report(Y_test,y_predict)
print(report)
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(Y_test,grid.predict_proba(X_test)[:,-1])
```
