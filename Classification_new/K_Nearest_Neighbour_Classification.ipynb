```python
import pandas as pd
```

```python
data=pd.read_csv('Social_Network_Ads.csv')
```

```python
data
```

```python
datanew=data.drop('User ID', axis=1)
```

```python
datanew
```

```python
datanew=pd.get_dummies(datanew,drop_first=True)
```

```python
datanew
```

```python
datanew.columns
```

```python
inde=datanew[['Age', 'EstimatedSalary',  'Gender_Male']]
```

```python
depn=datanew[['Purchased']]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(inde,depn,test_size=1/3,random_state=0)
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier=KNeighborsClassifier(n_neighbors=3)
classifier.fit(X_train,Y_train)
```

```python
y_predict=classifier.predict(X_test)
```

```python
y_predict
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(Y_test,y_predict)
cm
```

```python
from sklearn.metrics import classification_report
report=classification_report(Y_test,y_predict)
print(report)
```
