```python
import pandas as pd
```

```python
data=pd.read_csv('Social_Network_Ads.csv')
```

```python
data
```

```python
datanew=pd.get_dummies(data,drop_first=True)
```

```python
datanew
```

```python
from sklearn.model_selection import train_test_split
```

```python
datanew.columns
```

```python
datanew=datanew.drop('User ID',axis=1)
```

```python
datanew
```

```python
datanew.columns
```

```python
independant=datanew[['Age', 'EstimatedSalary', 'Gender_Male']]
```

```python
dependant=datanew[['Purchased']]
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(independant,dependant,test_size=1/3,random_state=0)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
classifier=DecisionTreeClassifier()
```

```python
classifier.fit(X_train,Y_train)
```

```python
y_predict=classifier.predict(X_test)
```

```python
y_predict
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm=confusion_matrix(Y_test,y_predict)
```

```python
cm
```

```python
from sklearn.metrics import classification_report
```

```python
clf_report=classification_report(Y_test,y_predict)
```

```python
clf_report
```

```python
print(clf_report)
```
