```python
import pandas as pd
```

```python
data=pd.read_csv('insurance_pre.csv')
```

```python
data
```

```python
data1=pd.get_dummies(data,drop_first=True)
```

```python
data1
```

```python
from sklearn.model_selection import train_test_split
```

```python
data1.columns
```

```python
independand=data1[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
```

```python
dependand=data1[['charges']]
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(independand,dependand,test_size=0.3,random_state=42)
```

```python
# from sklearn.svm import SVR
# regressor=SVR(kernel='linear',C=1000000)
# regressor.fit(X_train,Y_train)
```

```python
#criterion{“squared_error”, “absolute_error”, “friedman_mse”, “poisson”}, default=”squared_error”
#max_features{“sqrt”, “log2”, None}, int or float, default=1.0
from sklearn.ensemble import RandomForestRegressor
regressor=RandomForestRegressor(n_estimators=100, criterion='squared_error',max_features='sqrt',random_state=0)#max_features=None,
regressor.fit(X_train,Y_train)
```

```python
#criterion{“squared_error”, “friedman_mse”, “absolute_error”, “poisson”}, default=”squared_error”
#splitter{“best”, “random”}, default=”best”
#max_featuresint, float or {“sqrt”, “log2”}, default=None
# from sklearn.tree import DecisionTreeRegressor
# regressor=DecisionTreeRegressor(criterion='absolute_error',splitter='best',max_features=None)
# regressor.fit(X_train,Y_train)
```

```python
y_predict=regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r_score=r2_score(Y_test,y_predict)
```
