```python
from lightgbm import LGBMRegressor
```

```python
import pandas as pd
```

```python
data=pd.read_csv('insurance_pre.csv')
```

```python
datanew=pd.get_dummies(data,drop_first=True)
```

```python
datanew.columns
```

```python
indep=datanew[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
```

```python
depend=datanew[['charges']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(indep,depend,test_size=0.3,random_state=0)
```

```python
from lightgbm import LGBMRegressor
```

```python
regressor=LGBMRegressor()
regressor.fit(X_train,Y_train)
```

```python
y_predict=regressor.predict(X_test)
```

```python
y_predict
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(Y_test,y_predict)
```
