```python
import pandas as pd
```

```python
data=pd.read_csv('insurance_pre.csv')
```

```python
data1=pd.get_dummies(data,drop_first=True)
```

```python
#criterion{“squared_error”, “absolute_error”, “friedman_mse”, “poisson”}, default=”squared_error”
#max_features{“sqrt”, “log2”, None}, int or float, default=1.0

from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
# param_grid={'criterion':['squared_error','absolute_error','friedman_mse','poisson'],'max_features':['sqrt','log2',None],'n_estimators':[10,100]}
param_grid = {'criterion':['squared_error','absolute_error','friedman_mse','poisson'],'max_features': ['auto','sqrt','log2'],'n_estimators':[10,100]}
grid=GridSearchCV(RandomForestRegressor(),param_grid,refit=True,verbose=3,n_jobs=-1)
grid.fit(independend,dependend)
```
