```python
import pandas as pd
```

```python
data=pd.read_csv('insurance_pre.csv')
```

```python
data1=pd.get_dummies(data,drop_first=True)
```

```python
#kernel{‘linear’, ‘poly’, ‘rbf’, ‘sigmoid’, ‘precomputed’} 
#gamma{‘scale’, ‘auto’}
#C=1.0
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR
# params={'kernel':['linear', 'poly', 'rbf', 'sigmoid'],'C':[0.1,0.5,10,100],'gamma':['scale','auto']}
param_grid = {'kernel':['rbf','poly','sigmoid','linear'],
'C':[10,100,1000,2000,3000],'gamma':['auto','scale']}
grid=GridSearchCV(SVR(),param_grid,refit=True,verbose=3,n_jobs=-1)
grid.fit(independend,dependend)
```

```python
cv=grid.best_score_
```

```python
cv
```

```python
cv=grid.cv_results_
```

```python
cv
```

```python
table=pd.DataFrame.from_dict(cv)
```
