```python
import pandas as pd
```

```python
data=pd.read_csv('insurance_pre.csv')
```

```python
data1=pd.get_dummies(data,drop_first=True)
```

```python
data1
```

```python
from sklearn.model_selection import train_test_split
```

```python
data1.columns
```

```python
independend=data1[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
```

```python
dependend=data1[['charges']]
```

```python
independend
```

```python
dependend
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(independend,dependend,train_size=0.3,random_state=0)
```

```python
#kernel{â€˜linearâ€™, â€˜polyâ€™, â€˜rbfâ€™, â€˜sigmoidâ€™, â€˜precomputedâ€™} 
#gamma{â€˜scaleâ€™, â€˜autoâ€™}
#C=1.0
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV
# params={'kernel':['linear', 'poly', 'rbf', 'sigmoid'],'C':[0.1,0.5,10,100],'gamma':['scale','auto']}?
param_grid = {'kernel':['rbf','linear'],
'C':[0.1],'gamma':['auto','scale']}
grid=GridSearchCV(SVR(),param_grid,refit=True,verbose=3,n_jobs=-1)
grid.fit(X_train,Y_train)
```

```python
cv=grid.cv_results_
```

```python
cv
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
table=pd.DataFrame.from_dict(cv)
```
