```python
import pandas as pd
```

```python
data=pd.read_csv('insurance_pre.csv')
```

```python
data1=pd.get_dummies(data,drop_first=True)
```

```python
data1
```

```python
from sklearn.model_selection import train_test_split
```

```python
data1.columns
```

```python
independend=data1[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
```

```python
dependend=data1[['charges']]
```

```python
independend
```

```python
dependend
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(independend,dependend,train_size=0.3,random_state=0)
```

```python
#criterion{“squared_error”, “absolute_error”, “friedman_mse”, “poisson”}, default=”squared_error”
#max_features{“sqrt”, “log2”, None}, int or float, default=1.0

from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
param_grid={'criterion':['squared_error','absolute_error','friedman_mse','poisson'],'max_features':['sqrt','log2',None],'n_estimators':[10,100]}
grid=GridSearchCV(RandomForestRegressor(),param_grid,refit=True,verbose=3,n_jobs=-1)
grid.fit(X_train,Y_train)
```

```python
re=grid.cv_results_
```

```python
re
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
table=pd.DataFrame.from_dict(re)
```
