```python
import pandas as pd
```

```python
data=pd.read_csv('insurance_pre.csv')
```

```python
data
```

```python
datanew=pd.get_dummies(data,drop_first=True)
```

```python
datanew
```

```python
datanew.columns
```

```python
independand=datanew[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
```

```python
dependand=datanew[['charges']]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(independand,dependand,test_size=0.3,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train_pre=sc.fit_transform(X_train)
X_test_pre=sc.transform(X_test)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR
param_grid = {'kernel':['rbf','poly','sigmoid','linear'],
'C':[0.1],'gamma':['auto','scale']}
grid=GridSearchCV(SVR(),param_grid,refit=True,verbose=3,n_jobs=-1)
grid.fit(X_train_pre,Y_train)
```

```python
grid.best_score_
```

```python
cv=grid.cv_results_
```

```python
table=pd.DataFrame.from_dict(cv)
```

```python
table
```

```python
y_predict=grid.predict(X_test_pre)
```

```python
y_predict
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(Y_test,y_predict)
r_score
```
