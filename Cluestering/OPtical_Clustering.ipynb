```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('Mall_Customers.csv')
```

```python
datanew=data.copy()
```

```python
datanew
```

```python
X=data.iloc[:,[3,4]].values
```

```python
X
```

```python
from sklearn.cluster import OPTICS
SC=OPTICS(min_samples=2)
y_sc=SC.fit_predict(X)
```

```python
y_sc
```

```python
datanew['clusterdata']=y_sc
```

```python
datanew
```

```python
import seaborn as scb
scb=scb.lmplot(data=datanew,x=datanew.columns[3],y=datanew.columns[4],hue=datanew.columns[5],fit_reg=False)
```
