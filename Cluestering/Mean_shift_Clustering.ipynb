```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('Mall_Customers.csv')
```

```python
data
```

```python
X=data.iloc[:,[3,4]].values
```

```python
Y=X.copy()
```

```python
X
```

```python
from sklearn.cluster import MeanShift
MS=MeanShift(bandwidth=2).fit(X)
x_ms=MS.fit_predict(X)
```

```python
x_ms
```

```python
datanew=data.copy()
```

```python
datanew['clusterdat']=x_ms
```

```python
datanew
```

```python
import seaborn as scs
fact=scs.lmplot(data=datanew,x=datanew.columns[3],y=datanew.columns[4],hue=datanew.columns[5],fit_reg=False)
```
