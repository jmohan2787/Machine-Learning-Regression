```python
import pandas as pd
```

```python
data=pd.read_csv("50_Startups.csv")
```

```python
datanew=pd.get_dummies(data,drop_first=True)
```

```python
datanew.columns
```

```python
independent=datanew[['R&D Spend', 'Administration', 'Marketing Spend','State_Florida', 'State_New York' ]]
```

```python
dependand=datanew[['Profit']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(independent,dependand,test_size=0.3,random_state=0)
```

```python
#https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeRegressor.html
from sklearn.tree import DecisionTreeRegressor
```

```python
#criterion:{“squared_error”, “friedman_mse”, “absolute_error”, “poisson”}, default=”squared_error” & 
#splitter: {“best”, “random”}, default=”best”
#max_features: int, float or {“sqrt”, “log2”}, default=None
regressor=DecisionTreeRegressor(criterion='squared_error',splitter='best')
```

```python
regressor=regressor.fit(X_train,Y_train)
```

```python
import matplotlib.pyplot as plt
from sklearn import tree
tree.plot_tree(regressor)
plt.show()
```

```python
y_predict=regressor.predict(X_test)
```

```python
y_predict
```

```python
from sklearn.metrics import r2_score
```

```python
r_score=r2_score(Y_test,y_predict)
```
