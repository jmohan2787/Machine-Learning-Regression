```python
#Pandas is a library and its contains database related functions.
import pandas as pd
```

```python
# copy .csv file data intot 'data' variable
data=pd.read_csv('Salary_Data.csv')
```

```python
#copy all input column data into independant variable
independant=data[['YearsExperience']]
```

```python
# copy all output column data into dependant variable
dependant=data[['Salary']]
```

```python
independant
```

```python
dependant
```

```python
#import train_test_split model fucntion from sklearn.model_selection to splite train & test data set from copied variable.
from sklearn.model_selection import train_test_split
```

```python
# train adn test data split done for both X and Y axis
X_train,X_test,Y_train,Y_test=train_test_split(independant,dependant,test_size=1/3,random_state=0)
```

```python
# import library of LinearRegression from linear_model
from sklearn.linear_model import LinearRegression
```

```python
#copy all function of LinearRegression to regressor.
regressor=LinearRegression()
```

```python
# to create the model using train dataset
regressor.fit(X_train,Y_train)
```

```python
#copy the weight of the linear regression into weight variable
weight=regressor.coef_
```

```python
# read value of the weight
weight
```

```python
# copy the calcuated bias value from the created model into variable bias
bias=regressor.intercept_
```

```python
#read the biase variable
bias
```

```python
#predict the value for all splited test data
y_Predict=regressor.predict(X_test)
```

```python
# import the r2_scrore library to check r2 score by passing actual data & predicted data
from sklearn.metrics import r2_score
r_score=r2_score(Y_test,y_Predict)
```

```python
# read the score value & it should be nearly 1 is the good score.
r_score
```

```python
# import pickle library to copy the created model
import pickle
```

```python
# create the file name to take copy
filename='finalized_model_linear.sav'
```

```python
#take copy/write of model & save in 'finalized_model_linear.sav' this file.
pickle.dump(regressor,open(filename,'wb'))
```

```python
#load the created model in variable
load_model=pickle.load(open("finalized_model_linear.sav",'rb'))
```

```python
# predit/check with some random value 15
result=load_model.predict([[15]])
```
