```python
Quan=[i for i in dataset if dataset[i].dtype!='O']
Qual=[i for i in dataset if dataset[i].dtype=='O']
```

```python
print(Quan,Qual)
```

```python
table=pd.DataFrame(index=['Mean','Median','Mode','Q1:25%','Q2:50%','Q3:75%','89%','99%','Q4:100%'], columns=Quan)
```

```python
table
```

```python
dataset['sl_no'].mode()[0]
```

```python
import numpy as np
```

```python
for i in Quan:
    table[i]['Mean']=dataset[i].mean()
    table[i]['Median']=dataset[i].median()
    table[i]['Mode']=dataset[i].mode()[0]
    table[i]['Q1:25%']=dataset.describe()[i]['25%']
    table[i]['Q2:50%']=dataset.describe()[i]['50%']
    table[i]['Q3:75%']=dataset.describe()[i]['75%']
    table[i]['89%']=np.percentile(dataset[i],89)
    table[i]['99%']=np.percentile(dataset[i],99)
    table[i]['Q4:100%']=dataset.describe()[i]['max']
```
