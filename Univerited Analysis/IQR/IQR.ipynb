```python
Quan=[i for i in data.columns if data[i].dtype!="O"]
Qual=[i for i in data.columns if data[i].dtype=="O"]
```

```python
Quan
```

```python
updated_data=pd.DataFrame(index=['Mean','Median','Mode','Q1:25%','Q2:50%','Q3:75%','99%','Q4:100%','IQR','1.5Rule','LesserIQR','GreaterIQR','Min','Max'],columns=Quan)
```

```python
updated_data
```

```python
for i in updated_data:
    updated_data[i]['Mean']=data[i].mean()
    updated_data[i]['Median']=data[i].median()
    updated_data[i]['Mode']=data[i].mode()[0]
    updated_data[i]['Q1:25%']=data.describe()[i]['25%']
    updated_data[i]['Q2:50%']=data.describe()[i]['50%']
    updated_data[i]['Q3:75%']=data.describe()[i]['75%']
    updated_data[i]['99%']=np.percentile(data[i],99)
    updated_data[i]['Q4:100%']=data.describe()[i]['max']
    updated_data[i]['IQR']=updated_data[i]['Q3:75%']-updated_data[i]['Q1:25%']
    updated_data[i]['1.5Rule']=1.5*updated_data[i]['IQR']
    updated_data[i]['LesserIQR']=updated_data[i]['Q1:25%']-updated_data[i]['1.5Rule']
    updated_data[i]['GreaterIQR']=updated_data[i]['Q3:75%']+updated_data[i]['1.5Rule']
    updated_data[i]['Min']=data[i].min()
    updated_data[i]['Max']=data[i].max()
```

```python
updated_data
```

```python
LesserIQR=[]
GreaterIQR=[]
for i in updated_data:
    if updated_data[i]['Min'] < updated_data[i]['LesserIQR'] :
        LesserIQR.append(i)
    if updated_data[i]['Max'] > updated_data[i]['GreaterIQR']:
        GreaterIQR.append(i)
```
